```python
#Task-1:
my_list = [10,1,20,3,6,2,5]
for i in range (len(my_list)):
  for j in range (0,len(my_list)-i-1):
    if my_list[j]>my_list[j+1]:
      temp=my_list[j]
      my_list[j]=my_list[j+1]
      my_list[j+1]=temp
print(my_list)
```
