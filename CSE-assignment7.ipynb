```python
#task 4
def list1(list_a):  
    for i in range(len(list_a)):
        for j in range(i, len(list_a)):
            if i%2==0 and j%2==0:
                if list_a[i] > list_a[j]:
                    list_a[i], list_a[j] = list_a[j], list_a[i]             
    for k in range(len(list_a)):
        for l in range(k, len(list_a)):
            if k%2!=0 and l%2!=0:
                if list_a[k] < list_a[l]:
                    list_a[k], list_a[l] = list_a[l], list_a[k]
    return list_a             
print(list1([10,30,20,70,11,15,22,16,58,100,12,56,70,80]))
```

```python
#task5
name1=input('Enter name: ')
def sort1(list1,list2):
    for i in range(len(list1)-1,0,-1):
        for j in range(i):
            if list1[j]<list1[j+1]:
                list1[j],list1[j+1]=list1[j+1],list1[j]
                list2[j],list2[j+1]=list2[j+1],list2[j]
    return list1,list2
list1 = [ ["Alan", 95, 87, 91], ["Turing", 92, 90, 83], ["Elon", 87, 92, 80], ["Musk", 85, 94, 90] ]
name=["Alan","Turing","Elon","Musk"]
math=[91,83,80,90]
cse=[95,92,87,85]
phy=[87,90,92,94]
if name1=='MAT110':
    math,name_list=sort1(math,name)
    for i in name_list:
        print(i)
elif name1=='CSE110':
    cse,name_list=sort1(cse,name)
    for i in name_list:
        print(i)
elif name1=='PHY111':
    phy,name_list=sort1(phy,name)
    for i in name_list:
        print(i)
```
