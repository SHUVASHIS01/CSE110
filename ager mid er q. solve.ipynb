```python
#q1
n=input('')
d=0
v=0
c=0
s=0
for i in n:
  if i=='a' or i=='e' or i=='o' or i=='u' or i=='i':
    v+=1
  elif i=='0' or i=='1' or i=='2' or i=='3' or i=='4' or i=='5' or i=='6' or i=='7' or i=='8' or i=='9':
    d+=1
  elif i==' ':
    s+1
  else:
    c+=1
print('Number of digits',d)
print('Number of vowels',v)
print('Number of consonants',c)
```

```python
#q2
n=input('')
m=input('')
s=''
t=''
for i in n:
  i=int(i)
  if i%2==0:
    i=str(i)
    s+=i
  else:
    i=str(i)
    t+=i 
for j in m:
  j=int(j)
  if j%2==0:
    j=str(j)
    t+=j
  else:
    j=str(j)
    s+=j
print('First new number: ',s)
print('Second new number: ',t)
print('Subtraction of the newly generated numbers: ',int(s)-int(t))
```
