```python
#Task-14:
def function_name(sentence,position):
  s=''
  l=''
  for i in range(len(sentence)):
    if i==0:
      s+=sentence[i]
    elif i%position!=0:
      s+=sentence[i]
    else:
      l+=sentence[i]
  return s+l
sentence=input()
position=int(input())
print(function_name(sentence,position))
```

```python
#Task-18:
def function_name(starting_value,ending_value,first_divisor,second_divisor):
  sum=0
  for i in range(starting_value,ending_value):
    if i%first_divisor==0 and i%second_divisor==0:
      pass
    elif i%first_divisor==0 or i%second_divisor==0:
      sum+=i
  return sum
starting_value=int(input())
ending_value=int(input())
first_divisor=int(input())
second_divisor=int(input())
function_name(starting_value,ending_value,first_divisor,second_divisor)
```
