```python
#take a user input
#keep printing all the values from 0 to 100 unless user input is reached
#if the user input is not within 0 to 100 print 'no can do'
n1=int(input('input: '))
if(n1>=0 and n1<=100):
  for i in range(n1+1):
    print(i)
else:
  print('no can do')
```

```python
#task using break
n1=int(input('input: '))
if(n1>=0 and n1<=100):
  for i in range(n1+1):
    if i==n1:
      break
    print(i)
else:
  print('no can do')
```
