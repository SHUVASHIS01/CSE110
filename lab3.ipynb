```python
#task 16
n=input('')
m=input('')
s=''
if (m not in n and len(n)>3):
  print(n[1:-1])
elif len(n)<3:
  print(n)
else:
  for i in n:
    if i!=m:
      s+=i
  print(s)
```

```python
#task 17
n=input('')
m=input('')
s=''
for i in n:
  if i!=m:
    s+=i
print(s)
p=n[len(s)+1::]
t=''
for j in p:
  if j!=m:
    t+=j
print(t)
```
